#### [2번 문제] 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

- 다형성 이란 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 것이다.
- 연산자 오버로딩도 다형성을 지원하는 중요한 기술이다.
- 밑의 예제는 같은 멤버 함수 area가 다르게 반응하는 것을 보여준다.

```python
# -*- coding: utf-8 -*-
class Point:
    def area(self):
        print "점의 넓이는 0 입니다."
      
class Circle(Point):
    def __init__(self, r):
        self.r = r
    def area(self):
        cal = 3.14 * self.r * self.r
        print "원의 넓이는", cal ,"입니다."
      

class Sphere(Point):
    def __init__(self, r):
        self.r = r
    def area(self):
        cal = 4 * 3.14 * self.r * self.r
        print "구의 넓이는", cal ,"입니다."  
        
class Cylinder(Point):
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def area(self):
        cal = 3.14 * self.r * self.r * self.h
        print "원기둥의 넓이는", cal ,"입니다."       
        
 
point = Point()
circle = Circle(3)
sphere = Sphere(3)
cylinder = Cylinder(3, 3)

for each in (point, circle, sphere, cylinder):
    each.area()
```

#### [3번 문제] 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

    > \>\>\> c()  <br/>
    > \>\>\> d()  <br/>

    > [Count (step: 1)] 12 <br/>

  - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 관련 메소드를 추가하시오.
    > \>\>\> c = c + 5  <br/>
    > \>\>\> d = d - 5  <br/>
    > \>\>\> print c  <br/>
    > [Count (step: 1)] 17 <br/> 
    > \>\>\> print d  <br/>
    > [Count (step: 2)] 9 <br/>  
    
  - 요구사항 6. 다음과 같은 관계연산 (+, -)이 수행될 수 있도록 \_\_cmp\_\_() 메소드를 추가하시오.
    > \>\>\> print c > 10  <br/>
    > True <br/> 
    > \>\>\> print d > 10 <br/>
    > False <br/>
    > \>\>\> print c < 10  <br/>
    > False <br/> 
    > \>\>\> print d < 10 <br/>
    > True <br/>
    > \>\>\> print c == 10  <br/>
    > True <br/> 
    > \>\>\> print d != 9 <br/>
    > False <br/>

```python
# -*- coding: utf-8 -*-
class Counter():
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
                
    def __str__(self):
        result = "[Count (step: " + str(self.step) + ")] " + str(self.count)
        return result
    
    def incr(self):
        self.count = self.count + self.step
        
    def __call__(self):
        Counter.incr(self)
        
    def __add__(self, x):
        self.count = self.count + x
        return self
        
    def __sub__(self, x):
        self.count = self.count - x
        return self
    
    def __cmp__(self, other):
        return self.count - other
       
if __name__ == "__main__":         
    c = Counter(10)
    d = Counter(10, 2)
    
    print "요구사항 2 __str__()"
    print c
    print d
    
    print
    print "요구사항 3 incr()"
    c.incr()
    d.incr()
    print c
    print d
    
    print
    print "요구사항 4 __call__()"
    c()
    d()
    print c
    print d
    
    print
    print "요구사항 5 __add__(), __sub__()"
    c = c + 5
    print c
    d = d - 5
    print d
    
    print
    print "요구사항 6 __cmp__()"
    print c > 10
    print d > 10
    print c < 10
    print d < 10
    print c == 10
    print d != 9
```

- 요구사항 1: init 메소드를 구현한다. 2개의 인자를 받고 step은 디폴트 값을 준다. self를 이용하여 자신의 이름 공간에 변수를 저장한다.
- 요구사항 2: str 메소드를 구현한다. 자신의 이름공간에 저장된 변수를 알맞게 출력해준다.
- 요구사항 3: incr 메소드를 구현한다. 자신의 이름공간에 저장된 count를 step만큼 증가 시켜준다.
- 요구사항 4: call 메소드를 구현한다. 내부에서 이전에 구현한 incr 메소드를 호출한다.
- 요구사항 5: add, sub 메소드를 구현한다. 입력받은 x값을 이용하여 count값을 증가 감소 시킨다.

#### [4번 문제] 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str()\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
  - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

- init 은 생성자 메소드이다. 하나의 인자를 받고 반복문과 append를 이용하여 인자를 self에 저장한다. 그 후 elimicate_duplicate 메소드에 self를 인자로 넘겨준다.
- str 은 print 예약어나 str() 내장함수 호출에 대응되는 메소드이다. str이 호출되면 self에 저장된 값을 문자열 result에 저장하고 result를 반환한다.
- elimicate_duplicate 메소드는 입력받은 인자(리스트)의 중복되는 값을 제거해주는 역활을 한다. s라는 list를 객체를 선언하고 for ~ in 반복문과 if문, not in 을 이용하여 입력받은 인자 l의 중복되는 값을 제거하여 저장한다. 그리고 s를 이용하여 l에 다시 저장하는 과정을 통해 중복되는 값을 제거한다. elimicate_duplicate 메소드는 @staticmethod를 이용하여 정적 메소드로 구현되었기 때문에 인스턴스 객체와 무관하게 일반적으로 클래스 이름을 이용하여 직접 호출할 수 있는 메소드이다.

#### [5번 문제] 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오 (정답을 각 요구사항별로 입력할 필요 없이 5번 문제에 대해 1개의 MySet 클래스 정의 코드를 제시하면 된다.)
  - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    > \>\>\> s = s | t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {1 ,2 ,3 ,4 ,5 ,6 ,7 ,8 ,9} <br/> 
    
  - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> s = s & t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {3} <br/>

  - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> s = s - t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {1, 2} <br/>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
        
    def __or__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
        return self
        
    def __and__(self, l):
        s = []
        for e in l:
            if e in self:
                s.append(e)
        self[:] = []
        for e in s:
            self.append(e)
        MySet.eliminate_duplicate(self)
        return self
        
    def __sub__(self, l):
        s = []
        for e in self:
            if e not in l:
                s.append(e)
        self[:] = []
        for e in s:
            self.append(e)
        MySet.eliminate_duplicate(self)
        return self
        
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    
    print 
    print "요구사항 1 | 연산 합집합"
    s = s | t
    print s
    
    print 
    print "요구사항 2 & 연산 교집합"
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    s = s & t
    print s
    
    print 
    print "요구사항 3 - 연산 차집합"
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    s = s - t
    print s
```

- 요구사항 1: | 연산자를 구현하기위해 or 메소드를 구현한다. self에 피연산자 l을 추가하고 eliminate_duplicate을 이용하여 중복되는 값을 제거해준다.
- 요구사항 2: & 연산자를 구현하기위해 and 메소드를 구현한다.for ~ in 반복문과 if in을 이용하여 두 집합에 속해있는 값을 s에 저장한 후, self를 초기화 해주고 s에 저장된 값을 저장하여 교집합을 구현한다.
- 요구사항 3: - 연산자를 구현하기위해 sub 메소드를 구현한다. for ~ in 반복문과 if not in을 이용하여 self에만 속해 있는 값을 s에 저장하고 self를 초기화 해주고 s에 저장된 값을 저장한여 차집합을 구현한다.

```python
# -*- coding: utf-8 -*-
import urllib
import string
from collections import Counter

class WebWordsFrequency():
    def __init__(self, *args):
        self.url = []
        self.counts = {}
        for i in args:
            if i not in self.url:
                self.url.append(i)
        WebWordsFrequency.setWordsFrequency(self)
    
    def addUrl(self, url):
        if url not in self.url:
            self.url.append(url)
            WebWordsFrequency.setWordsFrequency(self)
        
    def removeUrl(self, url):
        if url in self.url:
            self.url.remove(url)
            WebWordsFrequency.setWordsFrequency(self)
    
    def listUrls(self):
        for i in self.url:
            print i
            
    def setWordsFrequency(self):
        self.counts.clear()
        for x in self.url:
            source = urllib.urlopen(x).read()
            len_source = len(source)
        
            i = 0
            while (i < len_source):
                found = False
                if (found != True and source[i] == "<" and source[i+1 : i+7].lower() == "script"):
                    found = True;
                    j = i + 7;
                    while (True):
                        if (source[j] == "<" and source[j+1] == "/" and source[j+2 : j+8].lower() == "script"):
                            source = source[:i] + ' ' + source[j+9:]
                            break
                        j = j + 1
                        if (j > len_source):
                            break
                    len_source = len(source)
                i = i + 1
            
            len_source = len(source)
            
            i = 0
            while (i < len_source):
                found = False
                if (found != True and source[i] == "<" and source[i+1 : i+6].lower() == "style"):
                    found = True;
                    j = i + 6;
                    while (True):
                        if (source[j] == "<" and source[j+1] == "/" and source[j+2 : j+7].lower() == "style"):
                            source = source[:i] + ' ' + source[j+8:]
                            break
                        j = j + 1
                        if (j > len_source):
                            break
                    len_source = len(source)
                i = i + 1
            
            len_source = len(source)
            
            i = 0
            while (i < len_source):
                found = False
                if (found != True and source[i] == "<"):
                    found = True;
                    j = i + 1;
                    while (True):
                        if (source[j] == ">"):
                            source = source[:i] + ' ' + source[j+1:]
                            break
                        j = j + 1
                        if (j > len_source):
                            break
                    len_source = len(source)
                i = i + 1
            
            words = ("".join(i for i in source if i not in string.punctuation)).split()
            
            temp = dict(Counter(words))
            for key in temp:
                if key in self.counts:
                    self.counts[key] += temp[key]
                else:
                    self.counts[key] = temp[key]
    
    def getWordsFrequency(self):
        if self.url == []:
            print self.url
            return None
        
        result = "{"
        for i in self.counts.keys():
                result = result +"'"+ str(i) + "': " + str(self.counts[i]) + ", "                 
        result = result[0:len(result)-2] + "}"
        
        print result
        
    def getMaxFreqencyWords(self):
        if self.url == []:
            print "None"
            return None
        
        getKey = []
        getValue = 0
        for key, value in self.counts.items():
            if(value == getValue):
                getKey.append(key)
                
            if(value > getValue):
                getKey[:] = []
                getKey.append(key) 
                getValue = value
            
        for i in getKey:
            print i, getValue

if __name__ == "__main__":
    w1 = WebWordsFrequency() 
    
    print "url이 등록되지 않은 w1객체 생성 "
    print ">>>w1.listUrls()"
    w1.listUrls()    
    print ">>>w1.getWordsFrequency()"
    w1.getWordsFrequency()
    print ">>>w1.getMaxFreqencyWords()"
    w1.getMaxFreqencyWords()
    
    w1.addUrl('http://www.naver.com')
    w1.addUrl('http://www.daum.net')
    w1.addUrl('http://cse.koreatech.ac.kr')
    
    print "3개의 Url을 등록이후"
    print ">>>w1.listUrls()"
    w1.listUrls()
    print ">>>w1.getWordsFrequency()"
    w1.getWordsFrequency()
    print ">>>w1.getMaxFreqencyWords()"
    w1.getMaxFreqencyWords()
    
    w1.removeUrl('http://www.daum.net')
    
    print "http://www.daum.net Url 삭제이후"
    print ">>>w1.listUrls()"
    w1.listUrls()
    print ">>>w1.getWordsFrequency()"
    w1.getWordsFrequency()
    print ">>>w1.getMaxFreqencyWords()"
    w1.getMaxFreqencyWords()
```

- 가변인자를 받는 생성자 메소드를 구현한다. 인자가 들어오면 리스트 self.url에 저장하고 setWordsFrequency 메소드를 호출하여 출현 빈도를 계산한다.
- addUrl 메소드를 구현하여 입력된 인자가 self.url에 없으면 저장하고, setWordsFrequency 메소드를 호출하여 출현 빈도를 다시 계산한다.
- removeUrl을 구현하여 입력된 인자가 self.url에 있는지 판단하고 있으면 제거 후 setWordsFrequency 메소드를 호출하여 출현 빈도를 다시 계산한다.
- listUrl메소드는 self.url에 저장된 값을 출력해준다
- getWordsFrequency 메소드는 setWordsFrequency 메소드에서 계산한 출현빈도(self.counts)를 출력해준다
- getMaxFreqencyWords 메소드는 setWordsFrequency 메소드에서 계산한 출현빈도(self.counts)를 for ~ in문을 이용하여 가장 많이 출현한 단어를 출력해준다.
- setWordsFrequency 메소드는 과제3 8번 문제를 수정한 것으로 HTML 태그 및 CSS와 Javascript를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하는 역활을 한다.

#### [8번 문제] 7번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
    - 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
    
    > w4 = OrderedWebWordsFrequency('http://www.daum.net', 'http://www.naver.com', 'http://www.google.co.kr')
    > w4.getWordsFrequency()
    > [('안녕', 12), ('우리', 11), ('다음', 8), ('world', 2), ('hello', 1), ('python', 1)]

```python
from WebWordsFrequency import *
import operator

class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self):
        sorted_x = sorted(self.counts.items(), key=operator.itemgetter(1), reverse=True)
        
        result = "["
        for i in sorted_x:
                result = result + "('" + str(i[0]) + "': " + str(i[1]) + "), "                 
        result = result[0:len(result)-2] + "]"
        
        print result
        
    
if __name__ == "__main__":
    w4 = OrderedWebWordsFrequency('http://www.naver.com', 'http://www.daum.net')
    print ">>>w4.getWordsFrequency()"
    w4.getWordsFrequency()
    print ">>>w4.getMaxFreqencyWords()"
    w4.getMaxFreqencyWords()
```

- WebWordsFrequency을 모듈로 받아오고 상속을 통해 OrderedWebWordsFrequency을 구현한다.
- getWordsFrequency를 오버라이드 하여 정렬 후 출력해준다.
- operator 모듈을 추가하고, sorted를 통해 self.counts.items()을 키값을 operator.itemgetter(1)을 이용하여 사전의 Value(int)를 하게하고 하고 reverse = True하여 내림차순으로 정렬한다.
